```python
'''
Question1
Create a function that takes three integer arguments (a, b, c) and returns the amount of integers which are of equal value.
Examples
equal(3, 4, 3) ➞ 2

equal(1, 1, 1) ➞ 3

equal(3, 4, 1) ➞ 0 
Notes
Your function must return 0, 2 or 3.
'''
```

```python
def equal(a,b,c):
  if a==b and b==c:
    return 3
  elif a==b or b==c or a==c:
    return 2
  else:
    return 0    
```

```python
equal(3, 4, 3)
```

```python
equal(1, 1, 1)
```

```python
equal(3, 4, 1)
```

```python
'''
Question2
Write a function that converts a dictionary into a list of keys-values tuples.
Examples
dict_to_list({
  "D": 1,
  "B": 2,
  "C": 3
}) ➞ [("B", 2), ("C", 3), ("D", 1)]

dict_to_list({
  "likes": 2,
  "dislikes": 3,
  "followers": 10
}) ➞ [("dislikes", 3), ("followers", 10), ("likes", 2)]
Notes
Return the elements in the list in alphabetical order.
'''
```

```python
def dict_to_list(d):
  res=[]
  for i in sorted(d):
    res.append((i,d[i]))
  return res    
```

```python
dict_to_list({
  "D": 1,
  "B": 2,
  "C": 3
})
```

```python
dict_to_list({
  "likes": 2,
  "dislikes": 3,
  "followers": 10
})
```

```python
'''
Question3
Write a function that creates a dictionary with each (key, value) pair being the (lower case, upper case) versions of a letter, respectively.
Examples
mapping(["p", "s"]) ➞ { "p": "P", "s": "S" }

mapping(["a", "b", "c"]) ➞ { "a": "A", "b": "B", "c": "C" }

mapping(["a", "v", "y", "z"]) ➞ { "a": "A", "v": "V", "y": "Y", "z": "Z" }
Notes
All of the letters in the input list will always be lowercase.
'''
```

```python
def mapping(lst):
  d={}
  for i in lst:
    d[i]=i.upper()
  return d  
```

```python
mapping(["p", "s"])
```

```python
mapping(["a", "b", "c"])
```

```python
mapping(["a", "v", "y", "z"])
```

```python
'''
Question4
Write a function, that replaces all vowels in a string with a specified vowel.
Examples
vow_replace("apples and bananas", "u") ➞ "upplus und bununus"

vow_replace("cheese casserole", "o") ➞ "chooso cossorolo"

vow_replace("stuffed jalapeno poppers", "e") ➞ "steffed jelepene peppers"
Notes
All words will be lowercase. Y is not considered a vowel.
'''
```

```python
def vow_replace(str,s):
  res_str=""
  vowel="AEIOUaeoiu"
  for i in str:
    if i in vowel:
      res_str=str.replace(i,s)
  return res_str     
```

```python
vow_replace("apples and bananas", "u")
```

```python
vow_replace("cheese casserole", "o") 
```

```python
vow_replace("stuffed jalapeno poppers", "e")
```

```python
'''
Question5
Create a function that takes a string as input and capitalizes a letter if its ASCII code is even and returns its lower case version if its 
ASCII code is odd.
Examples
ascii_capitalize("to be or not to be!") ➞ "To Be oR NoT To Be!"

ascii_capitalize("THE LITTLE MERMAID") ➞ "THe LiTTLe meRmaiD"

ascii_capitalize("Oh what a beautiful morning.") ➞ "oH wHaT a BeauTiFuL moRNiNg."
'''
```

```python
def ascii_capitalize(str):
  str1=""
  for i in str:
    if ord(i)%2==0:
      str1=str1+i.upper()
    else:
      str1=str1+i.lower()
  return str1      
```

```python
ascii_capitalize("to be or not to be!")
```

```python
ascii_capitalize("THE LITTLE MERMAID")
```

```python
ascii_capitalize("Oh what a beautiful morning.") 
```
