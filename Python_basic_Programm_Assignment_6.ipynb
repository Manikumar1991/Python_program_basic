```python
#1.Write a Python Program to Display Fibonacci Sequence Using Recursion?
def fib_recursion(n):
  if n<=1:
    return n
  else:
    return (fib_recursion(n-1) +fib_recursion(n-2))
```

```python
aterms = int(input("Enter the number : "))
if aterms <=0:
  print("Please enter a positive number :")   
else:
  print("The fibonacci series :")  
  for num in range (aterms):
    print(fib_recursion(num))
```

```python
#2.Write a Python Program to Find Factorial of Number Using Recursion?
def fact(n):
  if n<=0:
    return 1
  else:
    return n*fact(n-1)  
fact(5)      
```

```python
#3.Write a Python Program to calculate your Body Mass Index?
def BMI(weight,height):
  return (weight/(height**2))
BMI(75.8,165.8)  
```

```python
#4.Write a Python Program to calculate the natural logarithm of any number?
import math as m
def log_func(a,b): # log function with user defined number base "b"
  return m.log (a,b)
log_func(45,3)
```

```python
def natural_func(a):  # natural log function with base "e" as default
  return m.log (a)
natural_func(45)  
```

```python
#5.Write a Python Program for cube sum of first n natural numbers?
import math as m
def sum_cube(n):
  sum=0 # initilize
  for i in range(0,n+1):
    sum=sum+m.pow(i,3)
  return sum

sum_cube(3)
```
