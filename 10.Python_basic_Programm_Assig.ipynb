```python
#1.Write a Python program to find sum of elements in list?
def sum_ele(lst):
  sum=0
  for i in lst:
    try:
      sum=sum+i
    except Exception as e:
      print(e)
  return sum  
sum_ele([1,2,3,'python',5,6])      
```

```python
#2.Write a Python program to  Multiply all numbers in the list?
def mul_ele(lst):
  sum=1
  for i in lst:
    try:
      if (type(i)==int):
        sum=sum*i
    except Exception as e:
      print(e)
  return sum  
mul_ele([1,2,3,'python',5,6]) 
```

```python
#3.Write a Python program to find smallest number in a list?
def small_num(lst):
  lowest=1e100    # Initilize the number
  for i in lst:
    if i<lowest:
      lowest=i
  return lowest
small_num([9,2,3,4,5,6,7,7,8,999])
```

```python
#4.Write a Python program to find largest number in a list?
def max_num(lst):
  max=0
  for i in lst:
    if i>max:
      max=i
  return max    
max_num([9,2,3,4,5,6,7,7,8,999])
```

```python
#5.Write a Python program to find second largest number in a list?
def second_largenumber(lst):
  lst.sort(reverse=True)
  return "The second largest number in lst :" ,lst[1]

second_largenumber([1,2,3,3,4,5,6,7,566,45,56])  
```

```python
#6.Write a Python program to find N largest elements from a list?
def n_largenumber(lst,n):
  lst.sort(reverse=True)
  return "The",n, " largest numbers in lst are :" ,lst[0:n]
n_largenumber([22,34,16,72,89],2)  
```

```python
#7.	Write a Python program to print even numbers in a list?
def even(lst):
  for i in lst:
    if (i%2==0):
      print(i, " is the even number")
even([12,33,13,22,6,7,10,9])      
```

```python
#8.Write a Python program to print odd numbers in a List?
def odd(lst):
  for i in lst:
    if(i%2!=0):
      print(i,"is ths odd number")
odd([12,13,45,67,22,2])      
```

```python
#9.Write a Python program to Remove empty List from List?
def rem_empty(lst):
  emptylist=[]
  new_list=[]
  for i in lst:
    if i!=emptylist:
      new_list.append(i)
  return new_list
rem_empty([1,2,3,[],6,7,89,[]])      

```

```python
#11.Write a Python program to Count occurrences of an element in a list?
def occurances(lst):
  d={}
  for i in lst:
    if i not in d:
      d[i]=1
    else:
      d[i]=d[i]+1   
  return d
occurances([1,1,2,3,3,3,3,5,5,5,5,55,])      
```
