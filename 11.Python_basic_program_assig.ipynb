```python
#4.Write a Python to check if a given string is binary string or not?
def func4(binary_string):
  binary_values=binary_string.split()
  ascii_string=""
  for binary_value in binary_values:
    an_integer=int(binary_value,2)
    ascii_char=chr(an_integer)
    ascii_string=ascii_string + ascii_char
  return ascii_string
func4("01100001 01100010 01100011")    

```

```python
#5.Write a Python program to find uncommon words from two Strings?
def func5():
  l=[]
  str1=input()
  str2=input()
  str1=str1.split()
  str2=str2.split()
  for word in str1:
    if word not in str2:
      l.append(word)
  for word1 in str2:
    if word1 not in str1:
      l.append(word1)
  return l
func5()      
```

```python
#6.Write a Python to find all duplicate characters in string?
def func6():
  string=input()
  lst=[]
  lst1=[]
  for char in string:
    if char not in lst:
      lst.append(char)
    else:
      lst1.append(char)
  return set(lst1) 
func6()       
```

```python
#7.Write a Python Program to check if a string contains any special character.
import re
def func7():
  string=input()
  special_char=re.compile('[@_!#$%^&*()<>?/\|}{~:]')
  if (special_char.search(string)==None):
    print("This string does not contain any special characters")
  else:
    print("This string contains special characters")  
func7()    
```
